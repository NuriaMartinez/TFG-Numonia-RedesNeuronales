```python
import numpy as np
import keras
from keras import layers
from keras.datasets import mnist
```

## Preparación del modelo

```python
# Model / data parameters
num_classes = 10
input_shape = (28, 28, 1) # 

# Load the data and split it between train and test sets
#como definir donde esta cada carpeta train test y val (mirar codigo de kaggle)

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Scale images to the [0, 1] range
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
# Make sure images have shape (28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
print("x_train shape:", x_train.shape)
print(x_train.shape[0], "train samples")
print(x_test.shape[0], "test samples")


# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
#BORRAR
import os
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf

# Función para cargar las imágenes desde una carpeta
def cargar_imagenes(ruta):
    imagenes = []
    etiquetas = []
    
    for etiqueta in os.listdir(ruta):
        etiqueta_ruta = os.path.join(ruta, etiqueta)
        for imagen_nombre in os.listdir(etiqueta_ruta):
            imagen_ruta = os.path.join(etiqueta_ruta, imagen_nombre)
            imagen = Image.open(imagen_ruta)
            imagen = imagen.resize((128, 128))  # Ajusta el tamaño de la imagen según sea necesario
            imagen = np.array(imagen) / 255.0  # Normaliza los valores de píxeles
            imagenes.append(imagen)
            etiquetas.append(etiqueta)
    
    return np.array(imagenes), np.array(etiquetas)

# Ruta de la carpeta principal con las imágenes
ruta_principal = 'ruta/a/la/carpeta/principal/imagenes'

# Cargar las imágenes y etiquetas de entrenamiento, prueba y validación
x_train, y_train = cargar_imagenes(os.path.join(ruta_principal, 'train'))
x_test, y_test = cargar_imagenes(os.path.join(ruta_principal, 'test'))
x_val, y_val = cargar_imagenes(os.path.join(ruta_principal, 'validation'))

# Dividir los datos de entrenamiento en entrenamiento y validación si es necesario
# x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

# Definir el modelo de red neuronal utilizando TensorFlow o Keras
modelo = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compilar el modelo
modelo.compile(optimizer='adam',
               loss='binary_crossentropy',
               metrics=['accuracy'])

# Entrenar el modelo
modelo.fit(x_train, y_train, epochs=10, validation_data=(x_val, y_val))

# Evaluar el modelo con los datos de prueba
test_loss, test_acc = modelo.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)
```

## Construcción del modelo

```python
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"), #una unica neurina, sigmoide
    ]
)

model.summary()
```

## Entrenar el modelo

```python
batch_size = 128 #jugar un poco cn este parametro (potencias de 2) (cuanto mayor sea el valor menos tarda)
epochs = 5

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"]) #cambias loss

model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1) #tener en cuenta neumonia y no neumonia dentro de train... validation= carpeta validacion
```

## Evaluación del modelo 

```python
score = model.evaluate(x_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
